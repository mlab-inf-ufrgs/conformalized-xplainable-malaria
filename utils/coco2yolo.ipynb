```python
import json
import os
from pathlib import Path
from collections import defaultdict
```

```python
train_coco_json_path = Path('../data/processed/training_coco.json')
test_coco_json_path = Path('../data/processed/test_coco.json')

train_output_dir = Path("../data/processed/dataset/labels/train")
test_output_dir = Path("../data/processed/dataset/labels/test")

train_output_dir.mkdir(parents=True, exist_ok=True)
test_output_dir.mkdir(parents=True, exist_ok=True)

def process_coco_json(coco_json_path, output_dir):
    with open(coco_json_path, "r") as f:
        data = json.load(f)

    images = {img['id']: img for img in data['images']}

    for ann in data['annotations']:
        img_info = images[ann['image_id']]
        img_name = Path(img_info['file_name']).stem
        img_width = img_info['width']
        img_height = img_info['height']

        x, y, w, h = ann['bbox']

        x_center = (x + w / 2) / img_width
        y_center = (y + h / 2) / img_height
        w_norm = w / img_width
        h_norm = h / img_height

        class_id = ann['category_id']

        line = f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}\n"

        out_file = os.path.join(output_dir, img_name + ".txt")

        with open(out_file, "a") as f:
            f.write(line)

    print(f"YOLO labels saved in: {output_dir}")

process_coco_json(train_coco_json_path, train_output_dir)

process_coco_json(test_coco_json_path, test_output_dir)
```
